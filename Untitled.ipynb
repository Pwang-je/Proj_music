```python
data = df_3

def percentile(x, p):
    p_index = int(p * len(x))
    return sorted(x)[p_index]

def year_play(year):
    
    x = data[data['년월'] == year]['재생시간']
    print("\n1사분위수 OPEN: ",percentile(x, 0.25))
    print()
    x_2 = max(data[data['년월'] == year]['재생시간'])
    print('\n최대시간 HIGH: ', x_2)
    x_1 = min(data[data['년월'] == year]['재생시간'])
    print('\n최소시간 LOW: ', x_1)
    x = data[data['년월'] == year]['재생시간']
    print("\n3사분위수 CLOSE: ",percentile(x, 0.75))
    
    
    
#     x = data[data['년월'] == year]['재생시간']
#     print("중앙값 : ",percentile(x, 0.5))
    print()
    

year_play(2018)
```
